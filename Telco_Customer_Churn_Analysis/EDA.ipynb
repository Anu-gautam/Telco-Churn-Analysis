```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
df = pd.read_csv('Cleaned_Telco_Customer_Churn.csv')
```

```python

# Set aesthetic style
sns.set_style("whitegrid")

# 1. Churn Distribution
plt.figure(figsize=(6, 4))
df['Churn'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#66b3ff','#ff9999'])
plt.title('Overall Churn Distribution')
plt.ylabel('')
plt.savefig('churn_distribution.png')

plt.close()
```

```python

# 2. Churn by Contract Type
plt.figure(figsize=(8, 5))
contract_churn = pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100
contract_churn.plot(kind='bar', stacked=True, color=['#66b3ff','#ff9999'])
plt.title('Churn Rate by Contract Type')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right')
plt.savefig('churn_by_contract.png')
plt.close()
```

```python

# 3. Churn by Internet Service
plt.figure(figsize=(8, 5))
internet_churn = pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100
internet_churn.plot(kind='bar', stacked=True, color=['#66b3ff','#ff9999'])
plt.title('Churn Rate by Internet Service')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc='upper right')
plt.savefig('churn_by_internet.png')
plt.close()
```

```python

# 4. Tenure Distribution by Churn
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Churn'] == 'No']['tenure'], label='Stayed', fill=True, color='#66b3ff')
sns.kdeplot(df[df['Churn'] == 'Yes']['tenure'], label='Churned', fill=True, color='#ff9999')
plt.title('Tenure Distribution (Stayed vs Churned)')
plt.xlabel('Tenure (Months)')
plt.legend()
plt.savefig('tenure_distribution.png')
plt.close()
```

```python

# 5. Monthly Charges vs Churn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette=['#66b3ff','#ff9999'])
plt.title('Monthly Charges vs Churn')
plt.savefig('monthly_charges_churn.png')
plt.close()
```

```python
# Calculate some key statistics for the response
total_churn_rate = df['Churn'].value_counts(normalize=True)['Yes'] * 100
monthly_contract_churn = contract_churn.loc['Month-to-month', 'Yes']
fiber_optic_churn = internet_churn.loc['Fiber optic', 'Yes']

print(f"Total Churn Rate: {total_churn_rate:.2f}%")
print(f"Month-to-month Churn Rate: {monthly_contract_churn:.2f}%")
print(f"Fiber Optic Churn Rate: {fiber_optic_churn:.2f}%")
```
