```python
import pandas as pd

# Load the dataset
file_path = 'WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(file_path)

# Inspect the data
print("Initial Data Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())

# Check for unique values in categorical columns to identify inconsistencies
for col in df.select_dtypes(include=['object']).columns:
    if col != 'customerID':
        print(f"\nUnique values in {col}:")
        print(df[col].unique())
```

```python
# 1. Convert TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

```

```python
# 2. Check for newly created NaNs
missing_total_charges = df['TotalCharges'].isnull().sum()
print(f"Missing values in TotalCharges after conversion: {missing_total_charges}")

```

```python
# 3. Investigate the rows with missing TotalCharges
print("\nRows with missing TotalCharges:")
print(df[df['TotalCharges'].isnull()][['tenure', 'MonthlyCharges', 'TotalCharges']])

```

```python
# 4. Handle missing values: Since tenure is 0, TotalCharges should be 0 or dropped.
# Setting to 0 makes sense as they haven't been billed yet.
df['TotalCharges'] = df['TotalCharges'].fillna(0)

```

```python
# 5. Check for duplicates
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

```

```python
# 6. Final check on data types and nulls
print("\nFinal Data Info:")
print(df.info())
```

```python

# 7. Save the cleaned data
cleaned_file_name = 'Cleaned_Telco_Customer_Churn.csv'
df.to_csv(cleaned_file_name, index=False)
```
